```python
# Input data
import pandas as pd

data_Q2 = pd.read_csv('palm_ffb.csv')
data_Q2.head(10)
```

```python
# Exlude the Date column
data2 = data_Q2.iloc[:,1:]
data2.head(10)
```

```python
# Anlyze descriptive
data2.describe()
```

```python
# Analyze correlation
data2.corr()
```

```python
# Analyze correlation by graph
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(data2.corr(),annot=True,cmap='RdYlGn',linewidths=1) 
fig=plt.gcf()
fig.set_size_inches(12,8)
plt.show()
```

```python
vars2 = [var for var in data2.columns if data2[var].dtypes != 'O']
```

```python
# Analyze the distributions each columns

def analyze_dist2(data2, var):
    df2 = data2.copy()
    df2[var].hist(bins=20)
    plt.ylabel('count')
    plt.xlabel(var)
    plt.title(var)
    plt.show()
    
for var in vars2:
    analyze_dist2(data2, var)
```

```python
# Anlyze skewness and kurtosis
d2_skew = data2.skew() 
d2_kurt = data2.kurt()

frame = { 'Skewness': d2_skew, 'Kurtosis': d2_kurt } 
  
print(pd.DataFrame(frame))
```

Conclusion:
1. Correlation between columns Max_temp and Average_temp have the highest value which is 0.76 and represent by soft green colour
2. Correlation between columns SoilMoisture and Average_temp have the lowest value which is -0.65 and represent by dark red colour
3. The distribution of Min_temp is negative skewed
4. All kurtosis of each columns is negative kurtosis with kurtosis value less than 3
